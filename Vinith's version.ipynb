```python
import pandas as pd

Eversource = pd.read_csv('Eversource.csv')
Eversource.head()
```

```python
from geoid import acs
```

```python
ct = pd.read_csv('ct_cbgs.csv')
ct.head()
```

```python
ct = ct.drop(['Current asthma among adults aged >=18 years', 'Coronary heart disease among adults aged >=18 years', 'Cancer (excluding skin cancer) among adults aged >=18 years', 'Diagnosed diabetes among adults aged >=18 years', 'Physical health not good for >=14 days among adults aged >=18 years', ], axis = 1)
```

```python
ct.head()
```

```python
ct_refined = ct[['GEOID10','Median household income (% of state median household income)', 'Area Median Income (State or metropolitan)', 'zipname']].copy()
ct_refined.head()
```

```python
ct_merged = pd.read_csv('merged_data.csv')
ct_refined = ct_merged[['Median household income (% of state median household income)', 'Area Median Income (State or metropolitan)', 'zipname_y', 'ZIPCODE', 'Housing burden (percent)', 'Energy burden', 'Energy burden (percentile)', 'Individuals over 64 years old', 'Physical health not good for >=14 days among adults aged >=18 years', 'FEMA Risk Index Expected Annual Loss Score', 'Housing + Transportation Costs % Income for the Regional Typical Household (percentile)', 'Percent individuals age 25 or over with less than high school degree (percentile)', 'Persistent Poverty Census Tract', 'Percent of individuals < 200% Federal Poverty Line (percentile)', 'Life expectancy (years)', 'Outages']].copy()
ct_refined['Outage_rank'] = ct_refined.Outages.rank(pct = True)  
ct_refined.head()
```

```python
ct_refined = ct_refined.groupby('ZIPCODE').mean()
```

```python
Final_data = pd.merge(ct_refined, Eversource, on='ZIPCODE')
```

```python
Final_data.head()
```

```python
Final_data.to_csv('final_data.csv')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
corr = Final_data.corr()
plt.figure(figsize=(14,8))
plt.title('Overall Correlation', fontsize=20)
sns.heatmap(corr,annot=False,cmap='RdYlBu',linewidths=0.2,annot_kws={'size':20})
plt.show()
```

```python
corr.to_csv('correlation.csv')
```
